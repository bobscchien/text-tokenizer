```python
# define paths of vocabularies
en_vocab_path  = os.path.join(DIR_VOCAB, 'en_vocab.txt')
jp_vocab_path  = os.path.join(DIR_VOCAB, 'jp_vocab.txt')
zh_vocab_path  = os.path.join(DIR_VOCAB, 'zh_vocab.txt')

print(ZHON_PUNC)
```

## <b> Chinese <b/>

#### <b> Dataset - Classical Chinese<b/>

```python
### download & examine the content of datasets

builder = tfds.builder("wikipedia/20190301.zh-classical", data_dir=DIR_DATA)
builder.download_and_prepare(download_dir=DIR_DATA)
print(builder.info)
```

```python
### create the datasets & preprocess the raw data for tokenization

original_dataset, _, _ = setup_tfds_builder(builder, pcts=((100, 0), 0), as_supervised=False)
display(tfds.as_dataframe(original_dataset.take(30), builder.info))

dataset_zhc = original_dataset.map(lambda text:text['text']).map(preprocessors['zh'])
```

#### <b> Dataset - Traditional Chinese <b/>

```python
### download & examine the content of datasets

config = tfds.translate.wmt.WmtConfig(
  version='1.0.0',
  language_pair=("zh", "en"),
  subsets={
    tfds.Split.TRAIN: ["wikititles_v1"],
    tfds.Split.VALIDATION: ["newstest2018"]
  }
)

builder = tfds.builder("wmt_translate", data_dir=DIR_DATA, config=config)
builder.download_and_prepare(download_dir=DIR_DATA)
print(builder.info)
```

```python
### create the datasets & preprocess the raw data for tokenization

original_dataset, _, _ = setup_tfds_builder(builder, pcts=((100, 0), 0), as_supervised=True)
display(tfds.as_dataframe(original_dataset.take(30), builder.info))

# zh-classical dataset is also included
dataset = original_dataset.map(lambda zh,en:tf.py_function(func=cc_converter, inp=[zh], Tout=tf.string)). \
                           map(preprocessors['zh']).concatenate(dataset_zhc)
```

#### <b> Tokenizer - Traditional Chinese <b/>

```python
# tokenization
tokenizer, vocab = build_bert_tokenizer(zh_vocab_path, dataset, cjk=True, vocab_size=None, revocab=True)

# demonstration
demo_tokenizer(tokenizer, dataset, sample=10)
```

## <b> English <b/>

#### <b> Dataset - English <b/>

```python
### download & examine the content of datasets

config = tfds.translate.wmt.WmtConfig(
  version='1.0.0',
  language_pair=("zh", "en"),
  subsets={
    tfds.Split.TRAIN: ["newscommentary_v14"],
    tfds.Split.VALIDATION: ["newstest2018"]
  }
)

builder = tfds.builder("wmt_translate", config=config)
builder.download_and_prepare(download_dir=DIR_DATA)
print(builder.info)
```

```python
### create the datasets & preprocess the raw data for tokenization

original_dataset, _, _ = setup_tfds_builder(builder, pcts=((100, 0), 0), as_supervised=True)
display(tfds.as_dataframe(original_dataset.take(30), builder.info))

# zh-classical dataset is also included
dataset = original_dataset.map(lambda en,zh:en).map(preprocessors['en'])
```

#### <b> Tokenizer - English <b/>

```python
# Configuration
bert_vocab_params = dict(
    bert_tokenizer_params = {'lower_case':True}, 
    # Reserved tokens that must be included in the vocabulary
    reserved_tokens = RESERVED_TOKENS,
    # Arguments for `wordpiece_vocab.wordpiece_tokenizer_learner_lib.learn`
    learn_params={}, #{'lower_thresh':1, 'slack_ratio':0, 'max_unique_chars':5000}
)

# Tokenization
tokenizer, vocab = build_bert_tokenizer(en_vocab_path, dataset, bert_vocab_params=bert_vocab_params, 
                                        vocab_size=15000, batch_size=1024, revocab=True)

# Demonstration
demo_tokenizer(tokenizer, dataset, sample=10)
```

## <b> Export <b/>

```python
### Build

tokenizers = tf.Module()
tokenizers.en = CustomTokenizer(RESERVED_TOKENS, en_vocab_path)
tokenizers.zh = CustomTokenizer(RESERVED_TOKENS, zh_vocab_path)
tokenizers.zhc = CustomTokenizer(RESERVED_TOKENS, zh_vocab_path)

### Save

# Tokenizer
tokenizer_path = os.path.join(DIR_TOKEN, 'bert_tokenizers')

if os.path.isdir(tokenizer_path):
    shutil.rmtree(tokenizer_path)
tf.saved_model.save(tokenizers, tokenizer_path)

### Reload
tokenizers = tf.saved_model.load(tokenizer_path)

for lang in dir(tokenizers):
    func = getattr(tokenizers, lang)
    if hasattr(func, 'get_vocab_size'):
        print(f'{lang} Vocabulary Size :', func.get_vocab_size().numpy())
```
